```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
file_name='C:\Pumpkin_Seeds_Dataset.xlsx'
iris=pd.read_excel(file_name)

from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
iris['Class']=label.fit_transform(iris['Class'])
iris.head()
```

```python
iris.keys()
```

```python
# 인덱스를 생성합니다.
index = iris.keys()
index
```

```python
iris_data = iris[index]
iris_target = iris.Class

df_data = pd.DataFrame(data=iris_data)
df_target = pd.DataFrame(data=iris.Class, columns=['label'])
df_data
```

```python
iris_data = pd.concat([df_data, df_target], axis=1)
iris_data
```

```python
df_data.isna().sum()
```

```python
x = iris_data
x = x.drop(columns=['label'], axis=1)
x
```

```python
y = iris_data['label']
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
colormap = plt.cm.PuBu
plt.figure(figsize=(8, 8))
plt.title("Analyze Correlation of Features", y = 1.0, size = 16)
sns.heatmap(iris_data.astype(float).corr(), linewidths = 0.1, vmax = 1.0,
           square = True, cmap = colormap, linecolor = "black", annot = True, annot_kws = {"size" : 16})
```

```python
x=iris.drop('Class', axis=1)
y=iris['Class']
```

```python
print("Shape of x:", x.shape)
print("Shape of y:", y.shape)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC

clf_svm = SVC(random_state=0)
clf_svm.fit(x_train, y_train)

pred_svm = clf_svm.predict(x_test)

print("\n--- SVM Classifier ---")
print(accuracy_score(y_test, pred_svm))
print(confusion_matrix(y_test, pred_svm))
```

```python
# LR

from sklearn.linear_model import LogisticRegression

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(x_train, y_train)

pred_lr = clf_lr.predict(x_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

```python
# DT

from sklearn.tree import DecisionTreeClassifier

clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(x_train, y_train)

pred_dt = clf_dt.predict(x_test)

print ("\n--- Decision Tree Classifier ---")
print (accuracy_score(y_test, pred_dt))
print (confusion_matrix(y_test, pred_dt))
```

```python
# RT

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

print ("\n--- Random Forest ---")
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(x_train, y_train)
pred = rf_clf.predict(x_test)
print(accuracy_score(y_test,pred))
print (confusion_matrix(y_test, pred))
```
